```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import control.matlab as control
import control as ct
```

```python
# create a transfer function system
G = ct.tf([0, 1], [1, 1])
print(G)
```

```python
# step response
y, t = ct.step_response(G)
plt.plot(t, y)
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Step Response')
```

```python
# impulse response
y, t = ct.impulse_response(G)
plt.plot(t, y)
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Step Response')
```

```python
# create a transfer function system
G2 = ct.tf([0, 1], [1, 10])
print(G2)
```

```python
# step response
y1, t1 = ct.step_response(G, T=np.linspace(0, 10, 1000), transpose=True)
plt.plot(t1, y1)

y2, t2 = ct.step_response(G2, T=np.linspace(0, 10, 1000), transpose=True)
plt.plot(t2, y2)

plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Step Response')
```

```python
# step response
y1, t1 = control.step(G, T=np.linspace(0, 10, 1000))
plt.plot(t1, y1)

y2, t2 = control.step(G2, T=np.linspace(0, 10, 1000))
plt.plot(t2, y2)

plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Step Response')
```

```python
# impulse response
y1, t1 = ct.impulse_response(G, T=np.linspace(0, 10, 1000))
plt.plot(t1, y1)

y2, t2 = ct.impulse_response(G2, T=np.linspace(0, 10, 1000))
plt.plot(t2, y2)
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Step Response')
```
